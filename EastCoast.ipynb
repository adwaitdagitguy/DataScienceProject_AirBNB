```python
# make necessary imports for a data science project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the file
# stratified sample file
data=pd.read_csv('/content/EastCoast.csv')
# remove features: listing_url, scrape_id, source, picture_url, host_url, host_thumbnail_url, host_picture_url, neighbourhood
data.drop(['listing_url', 'scrape_id', 'source', 'picture_url', 'host_url', 'host_thumbnail_url', 'host_picture_url', 'neighbourhood', 'City'], axis=1, inplace=True)
data.describe()
data.info()
```

```python
#  let us first handle the floats and int64 for missing values
# to do that create a list of such dtype columns
# also print the number of missing value ratio for each columne (assuming 5487 total recs)
dataNumericList=data.select_dtypes(include=['float64']).columns
print(dataNumericList)
data[dataNumericList].isnull().sum()/5487 * 100
```

```python
# drop the calendar_updated as it is totally null column
data.drop(['calendar_updated'], axis=1, inplace=True)
```

```python
floatColList=['review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value','reviews_per_month']
# Structural inaccuracy: bathrooms, bedrooms, beds as they contain only discrete values
# Imputation : replace with outlier, median and mean and check the differences in statistical parameters : viz mean, median, stddev
# Outlier treatment : replace
dataWithMeanReplacement=data.copy()
dataWithMedianReplacement=data.copy()
dataWithOutlierReplacement=data.copy()

# fill na of dataWithMeanReplacement  with the mean for floatColList
dataWithMeanReplacement[floatColList]=dataWithMeanReplacement[floatColList].fillna(dataWithMeanReplacement[floatColList].mean())
# fill na of dataWithMedianReplacement with the median for floatColList
dataWithMedianReplacement[floatColList]=dataWithMedianReplacement[floatColList].fillna(dataWithMedianReplacement[floatColList].median())
# fill na of dataWithOutlierReplacement with the positive extreme outlier i.e. mean + 3* std deviation
dataWithOutlierReplacement[floatColList]=dataWithOutlierReplacement[floatColList].fillna(dataWithOutlierReplacement[floatColList].mean() + 3*dataWithOutlierReplacement[floatColList].std())
```

```python
# LOAD mean, std, 50% keys in relevant data structure for 3 models
# ASSUME A NAIVE model, the model which returns maximum number of minimum keys is the best
infodic = {}
for model_name, model_df in {
    'Median Replacement': dataWithMedianReplacement,
    'Mean Replacement': dataWithMeanReplacement,
    'Outlier Replacement': dataWithOutlierReplacement
}.items():
    infodic[model_name] = {
        'mean': model_df.describe().loc['mean'],
        'std': model_df.describe().loc['std'],
        '50%': model_df.describe().loc['50%']
    }
# MEAN:
# MEDIAN:
# STDDEV:
def find_model_with_most_min_keys(infodic):
    """
    Finds the model with the most keys of minimum absolute value.
    Args:
        infodic: A dictionary containing statistics for different models.
    Returns:
        The name of the model with the most keys of minimum absolute value.
    """

    min_key_counts = {}
    for model_name, stats in infodic.items():
        min_keys_count = 0
        for stat_type, values in stats.items():
            min_val = np.min(np.abs(values))  # Find minimum absolute value
            min_keys = values[np.isclose(np.abs(values), min_val)]  # Get keys with minimum absolute value
            min_keys_count += len(min_keys)  # Count the number of minimum keys

        min_key_counts[model_name] = min_keys_count

    # Find the model with the maximum number of minimum keys
    best_model = max(min_key_counts, key=min_key_counts.get)
    return best_model

# Example usage:
best_model = find_model_with_most_min_keys(infodic)
print(f"The model with the most keys of minimum absolute value is: {best_model}")

```

```python
# Proceed with median replacement
# now it is time for handling the beds, bedrooms, bathrooms
# first create a separate list for them
rooms=['bedrooms','bathrooms','beds']
# print their unique values in the dataset
for room in rooms:
  print(dataWithMedianReplacement[room].unique())

# Although it is float 64, these are discrete values so we treat them categorically
# Methods : median replacement, mode replacement, replace with low frequency arbitrary values
dataModeReplacement=dataWithMedianReplacement.copy()
dataModeReplacement[rooms]=dataModeReplacement[rooms].fillna(dataModeReplacement[rooms].mode().iloc[0])

dataLowFreqReplacement=dataWithMedianReplacement.copy()
dataLowFreqReplacement[rooms]=dataLowFreqReplacement[rooms].fillna(dataLowFreqReplacement[rooms].mode().iloc[0])

dataMedianRoomsReplacement=dataWithMedianReplacement.copy()
dataMedianRoomsReplacement[rooms]=dataMedianRoomsReplacement[rooms].fillna(dataMedianRoomsReplacement[rooms].median())
```

```python
# create a new infodic and find the best model using definitions already defined
infodic = {}
for model_name, model_df in {
    'Mode Replacement': dataModeReplacement,
    'Low Frequency Replacement': dataLowFreqReplacement,
    'Median Replacement': dataMedianRoomsReplacement
}.items():infodic[model_name] = {
        'mean': model_df.describe().loc['mean'],
        'std': model_df.describe().loc['std'],
        '50%': model_df.describe().loc['50%']
    }

best_model = find_model_with_most_min_keys(infodic)
print(f"The model with the most keys of minimum absolute value is: {best_model}")
```

```python
# Remove structural inaccurcies: Fix Date some have yyyy-mm-dd others have mm-dd-yyyy
def format_date(date_str):
    try:
        # Attempt to parse as yyyy-mm-dd
        return pd.to_datetime(date_str, format='%Y-%m-%d').strftime('%d-%m-%Y')
    except ValueError:
        try:
            # If that fails, parse as mm/dd/yyyy
            return pd.to_datetime(date_str, format='%m/%d/%Y').strftime('%d-%m-%Y')
        except ValueError:
            # If both fail, return the original string (or handle as needed)
            return date_str

# Apply the formatting function to the date columns
for col in ['first_review', 'last_review', 'host_since']:
    dataDuplicatesHandled[col] = dataDuplicatesHandled[col].apply(format_date)
```
