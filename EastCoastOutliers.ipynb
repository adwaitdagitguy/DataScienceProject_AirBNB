```python
# prompt: number_of_reviews_ltm
# minimum_minimum_nights
# maximum_minimum_nights
# maximum_maximum_nights
# minumum_maximum_nights
# minimum_nights_avg_ntm
# maximum_nights_avg_ntm
# drop these columns

# Drop specified columns
columns_to_drop = ['number_of_reviews_ltm', 'minimum_minimum_nights', 'maximum_minimum_nights',
                   'maximum_maximum_nights', 'minimum_maximum_nights', 'minimum_nights_avg_ntm',
                   'maximum_nights_avg_ntm']
dataCleaned = dataCleaned.drop(columns=columns_to_drop, errors='ignore')
```

```python
# prompt: An outlier is defined using the 3 sigma rule. As the data is roughly normal, remove outliers for these columns:
# host_response_rate_in_percent
# host_acceptance_rate_in_percent
# host_listings_count
# host_total_listings_count
# price_in_dollars
# maximum_nights
# minimum_nights
# review_per_month

# Define a function to remove outliers based on the 3 sigma rule
def remove_outliers_3sigma(df, col):
    # Calculate the mean and standard deviation
    mean = df[col].mean()
    std = df[col].std()

    # Define the lower and upper bounds
    lower_bound = mean - 3 * std
    upper_bound = mean + 3 * std

    # Filter the DataFrame to remove outliers
    df_filtered = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    return df_filtered

# Specify the columns to process
columns_to_process = ['host_response_rate_in_percent', 'host_acceptance_rate_in_percent',
                      'host_listings_count', 'host_total_listings_count', 'price_in_dollars',
                      'maximum_nights', 'minimum_nights', 'reviews_per_month']

# Process each column
for col in columns_to_process:
    dataCleaned = remove_outliers_3sigma(dataCleaned, col)

print(dataCleaned.info())
```

```python
dataCleaned.drop(['neighbourhood_group_cleansed'], axis=1, inplace=True)
dataCleaned.info()
```
